```python
from ultralytics import YOLO
import cv2

# Load the YOLO model
model = YOLO('best.pt')

# Open the webcam
cap = cv2.VideoCapture(0)  # '0' is the default webcam

while cap.isOpened():
    ret, frame = cap.read()  # Read a frame from the webcam
    if not ret:
        break  # Break if the frame couldn't be captured

    # Perform prediction on the frame
    results = model.predict(source=frame, show=False, conf=0.5)  # Run detection without showing an automatic window

    # Visualize the results on the frame
    annotated_frame = results[0].plot()  # Annotate the frame with predictions

    # Display the annotated frame
    cv2.imshow("YOLO Detection", annotated_frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
cap.release()
cv2.destroyAllWindows()
```
